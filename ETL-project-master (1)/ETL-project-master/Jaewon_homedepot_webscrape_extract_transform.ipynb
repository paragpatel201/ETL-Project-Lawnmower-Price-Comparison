```python
from splinter import Browser
browser = Browser('chrome')
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.homedepot.com/s/lawn%2520mower?NCNI-5'
browser.visit(url)
```

```python
titles=[]
price=[]
images=[]
ratings=[]
links=[]
reviews=[]
page=24
 

for x in range(2,13):
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    results=soup.find_all('div',class_="js-pod")
    
    
    for result in results:
        try:
            title=result.find('input',class_="checkbox-btn__input")['data-title']
            titles.append(title)
            list_price=result.find('input',class_="checkbox-btn__input")['data-price']
            price.append(list_price)
            #original_price=result.find('input',class_="checkbox-btn__input")['data-was-price']
            image=result.find('input',class_="checkbox-btn__input")['data-img']
            images.append(image)
            rating=result.find('span',class_="stars")['rel']
            ratings.append(rating)
            link=result.find('a',class_="js-podclick-analytics")["href"]
            links.append(link)
            #model_number=result.find('div',class_="pod-plp__model").get_text().strip()
            review=result.find('div',class_="js-podclick-analytics pod-plp__ratings").get_text().strip()
            reviews.append(review)
            if (title and list_price and image and rating and link and model_number and review):
                print('-------------')
                print(title)
                print(list_price)
                print(image)
                print(rating)
                print(link)
                print(model_number)
                print(review)                    
        except AttributeError as e:
             print(e)
       
    s="/b/Outdoors/N-5yc1vZbx82/Ntk-ProductInfoMatch/Ntt-lawn%2Bmower?NCNI-5&amp;isSearch=true&amp;Ns=None&amp;Nao+str(page)" 
    
    browser.links.find_by_href(s)
    time.sleep(10)
    page +=24
    
    
   
```

```python
homepage="https://www.homedepot.com"
for link in links:
    l=homepage+link
    print(links)
```

```python
homedepot_data={"title":titles,"price":price,"imageUrl":images,"pageUrl":l,"rating":ratings,"totalReview":reviews}
```
