```python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
executable_path = {'executable_path': 'chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)

import pandas as pd
import re
```

```python
url = 'https://www.walmart.com/search/?page=1&ps=40&query=lawn+mower&typeahead=lawn'
browser.visit(url)
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
html = browser.html
soup = bs(html, "html.parser")
print(soup.prettify())
```

```python
results = soup.find_all('li', class_="Grid-col")
print(results)
```

```python
title_list = []
price_list = []
link_list = []
rating_list =[]
picture_list =[]

# Loop through returned results# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('a', class_="product-title-link").text
        
        # Identify and return price of listing
        price = result.find('span', class_="price-characteristic").text
        
        # Identify and return link to listing
        link = result.a["href"]
        
        # Identify and return ratings of listing
        rating = result.find('div', class_="stars").text
        
        # Identify and return IMAGE of listing
        image = result.find('img').get('data-image-src')
        

        # Print results only if title, price, and link are available
        if (title and price and link and rating and image):
            print('-------------')
            print(title)
            title_list.append(title)
            print(price)
            price_list.append(price)
            print(link)
            link_list.append("https:www.walmart.com"+link)
            print(rating)
            rating_list.append(rating)
            print(f'image: {image}')
            picture_list.append(image)
            
            
    except AttributeError as e:
        print(e)
```
