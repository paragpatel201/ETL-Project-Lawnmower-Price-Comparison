```python
# Dependencies
from bs4 import BeautifulSoup
#import requests
#from selenium import webdriver
from splinter import Browser
from bs4 import BeautifulSoup
import re 
import time
import pandas as pd
```

```python
# URL of page to be scraped
url = 'https://www.lowes.com/search?searchTerm=lawn+mower'
```

```python
# Retrieve page with the requests module
#headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}

#response = requests.get(url,headers=headers)
executable_path = {'executable_path': 'C:/Users/ssjaiswal/Documents/web_drivers/chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)

browser.visit(url)
time.sleep(20)
```

```python
brand=[]
reviews=[]
rating=[]
link=[]
image=[]
price=[]

for page in range(2,6):

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    #print(soup)
    #product-wrapper v-spacing-large art-pl-product-1-non-pvs position-relative
    #results = soup.find_all('li', class_=re.compile("product-wrapper v-spacing-large art-pl-product-.*-non-pvs position-relative$"))
    results = soup.find_all('li', class_=re.compile("product-wrapper"))
    print(results)
#brand, size, power_type, price, reviews, link, strength, source,rating
    for result in results:
        
        try:
            #print(result.find('li')['data-productid'])
            brand.append(result.get('data-producttitle'))
            reviews.append(result.get('data-productreviews'))
            rating.append(result.get('data-productrating'))
            link.append("https://www.lowes.com/"+result.get('data-producturl'))
            image.append(result.get('data-productimg'))

            #print(result.find_all('span'))
            #print("price: ")
            price.append(result.find_all('span')[7].text)
            #comment for now this is for sale price
                #lowes_dic["sale_price"]=result.find_all('span')[8].text
                #lowes_dic["sale_per"]=result.find_all('span')[9].text
            #    
            #print(a)
            #print(result.find('span',class_="green display-block v-spacing-mini small").text)
            #print("***********")
            
        except:
            print("tag not found!!!")
    
    s= "/search?searchTerm=lawn mower&page="+str(page)
    print(s)
    browser.click_link_by_href(s)  
    time.sleep(25)
  
```

```python
browser.quit()
```

```python
lowes_pd=pd.DataFrame({
"title":brand,
"price":price,
"imageUrl":image,
"pageUrl":link,
"rating":rating,
"totalReview":reviews,
"source":'lowes'   
      })

lowes_pd.head()
```
