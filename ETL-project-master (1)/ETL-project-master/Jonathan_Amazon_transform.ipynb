```python
import pandas as pd
```

```python
mower_csv = "amazon_lawn_mower_raw.csv"
mower_pd = pd.read_csv(mower_csv)
mower_pd
```

```python
mower_pd['source'] = 'Amazon'
```

```python
mower_cleaned = mower_pd[['title', 'price', 'imageUrl', 'detailPageURL', 'rating', 'totalReviews', 'source']].copy()
mower_cleaned = mower_cleaned.rename(columns={"detailPageURL":"pageUrl", "totalReviews":"totalReview"})

mower_cleaned = mower_cleaned.dropna()
mower_cleaned
```

```python
mower_cleaned.to_csv("amazon_lawn_mower_cleaned.csv", index=False, header=True)
```
